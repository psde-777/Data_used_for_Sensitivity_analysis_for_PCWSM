```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.cm as cm

def area_under_curve(x, y):
    return np.trapz(y, x)

# Load the data files
data_files = ['expe_data/expe_saccharification_high_glc.txt', 
             'expe_data/expe_saccharification_medium_glc.txt',
             'expe_data/expe_saccharification_low_glc.txt']

data_files2 = ['expe_data_Liu/expe_saccharification_high_glc.txt', 
             'expe_data_Liu/expe_saccharification_medium_glc.txt',
             'expe_data_Liu/expe_saccharification_low_glc.txt']

# Load the data files
data_files3 = ['Bura_high/new/Output/mean_saccharification_Bura_high.txt', 
             'Bura_medium/Output/mean_saccharification_Bura_medium.txt',
             'Bura_low/Output/mean_saccharification_Bura_low.txt']

data_files4 = ['Liu_high/Output/mean_saccharification_Liu_high.txt', 
             'Liu_medium/Output/mean_saccharification_Liu_medium.txt',
             'Liu_low/Output//mean_saccharification_Liu_low.txt']

new = np.loadtxt('Bura_high/new/Output/mean_saccharification_Bura_high.txt')

PlusBh = np.loadtxt('Bura_high/Plus05/Output/mean_saccharification_Bura_high.txt')
PlusBm = np.loadtxt('Bura_medium/Plus05/Output/mean_saccharification_Bura_medium.txt')
PlusBl = np.loadtxt('Bura_low/Plus05/Output/mean_saccharification_Bura_low.txt')

PlusLh = np.loadtxt('Liu_high/Plus05/Output/mean_saccharification_Liu_high.txt')
PlusLm = np.loadtxt('Liu_medium/Plus05/Output/mean_saccharification_Liu_medium.txt')
PlusLl = np.loadtxt('Liu_low/Plus05/Output/mean_saccharification_Liu_low.txt')

MinusBh = np.loadtxt('Bura_high/Minus0.5/Output/mean_saccharification_Bura_high.txt')
MinusBm = np.loadtxt('Bura_medium/Minus05/Output/mean_saccharification_Bura_medium.txt')
MinusBl = np.loadtxt('Bura_low/Minus05/Output/mean_saccharification_Bura_low.txt')

MinusLh = np.loadtxt('Liu_high/Minus05/Output/mean_saccharification_Liu_high.txt')
MinusLm = np.loadtxt('Liu_medium/Minus05/Output/mean_saccharification_Liu_medium.txt')
MinusLl = np.loadtxt('Liu_low/Minus05/Output/mean_saccharification_Liu_low.txt')

# Initialize lists to store the data points, curve parameters, and curve colors
color_palette = cm.tab10.colors
color_palette2 = cm.tab20.colors
data_points = []
data_points2 = []
data_points3 = []
data_points4 = []
curve_colors = [color_palette[0], color_palette[1], color_palette[2]]
curve_colors2 = [color_palette2[1], color_palette2[3], color_palette2[5]]

# Process each data file
for i, file_name in enumerate(data_files):
    # Load data from file
    data = np.loadtxt(file_name)
    x_data = data[:, 0]
    y_data = data[:, 1]

    # Store data points
    data_points.append((x_data, y_data))

for i, file_name in enumerate(data_files2):
    # Load data from file
    data2 = np.loadtxt(file_name)
    x_data2 = data2[:, 0]
    y_data2 = data2[:, 1]
    
    # Store data points
    data_points2.append((x_data2, y_data2))

for i, file_name in enumerate(data_files3):
    # Load data from file
    data3 = np.loadtxt(file_name)
    x_data3 = data3[:, 0]
    y_data3 = data3[:, 1]

    # Store data points
    data_points3.append((x_data3, y_data3))

for i, file_name in enumerate(data_files4):
    # Load data from file
    data4 = np.loadtxt(file_name)
    x_data4 = data4[:, 0]
    y_data4 = data4[:, 1]

    # Store data points
    data_points4.append((x_data4, y_data4))

```

```python


# Plotting
fig, ax = plt.subplots(figsize=[12,8])

# Plot data points and curves
for i, (x, y) in enumerate(data_points):
    label = ['High Bura', 'Medium Bura', 'Low Bura'][i]
    ax.scatter(x, y, label=label, edgecolor=curve_colors[i], marker=['o', 's', '^'][i], facecolor=curve_colors[i], linewidth=2, s=80)

for i, (x, y) in enumerate(data_points3):
    # Plot the curve
    #label = ['simulated High Bura', 'simulated Medium Bura', 'dimulated Low Bura'][i]
    ax.plot(x, y, color=curve_colors[i], linewidth=3, linestyle='-')

for i, (x, y) in enumerate(data_points2):
    label = ['160C Liu', '140C Liu', '180C Liu'][i]
    ax.scatter(x, y, label=label, edgecolor=curve_colors[i], marker=['o', 's', '^'][i], facecolor='none', linewidth=2, s=100)
    
for i, (x, y) in enumerate(data_points4):    
    # Plot the curve
    #label = ['simulated 160 Liu', 'simulated 140 Liu', 'simulated 180 Liu'][i]
    ax.plot(x, y, color=curve_colors[i], linewidth=3, linestyle='dotted')

# Add legend and labels
legend = ax.legend(fontsize=14)
ax.set_xlabel('Time [h]', fontsize=18)
ax.set_ylabel('Glucose conversion [%]', fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=15)

# Show the plot
plt.savefig('experimental_curves_Liu_and Bura.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.patches as mpatches
import matplotlib.colors as mcolors

color_palette = cm.tab20.colors
original_colors = plt.cm.tab20.colors

# Define a factor to darken the colors (e.g., 0.7 for 70% darkness)
darken_factor = 0.6

# Create a list to store the darker colors
darker_colors = []

# Darken each color in the original palette
for color in original_colors:
    darker_color = mcolors.to_rgb(color)
    darker_color = tuple(c * darken_factor for c in darker_color)
    darker_colors.append(darker_color)

def plot_function(a, b):
    x = np.linspace(-10, 15, 100)  # Range of x values
    y = a * x / (x + b)  # Calculate y values

    plt.plot(x, y, color='white', linewidth=5)  # Plot the curve

    # Mark the area around the curve
    plt.fill_between(x, y - 30, y + 30, color=color_palette[0])
    plt.fill_between(x, y + 30, 100, color=color_palette[16])
    plt.fill_between(x, -200, y - 30, color=color_palette[18])
    plt.plot(x, y - 30, color='black', linewidth=1)
    plt.plot(x, y + 30, color='black', linewidth=1)

    # Add text annotations
    plt.text(2.5, -94, 'EXPERIMENTAL DATA', color='white', ha='center', fontsize=30, weight='bold')
    plt.hlines(-85, -7, -4.5, color='white', linewidth=4)
    plt.text(0, 65, 'FAST DIGESTION', color='black', ha='center', fontsize=30, weight='bold')
    plt.text(9, 12, 'Y<150', color='black', ha='center', fontsize=26, weight='bold')
    plt.text(0, -160, 'SLOW DIGESTION', color='black', ha='center', fontsize=30, weight='bold')

    plt.xticks([])
    plt.yticks([])
    plt.ylim(-200,100)
    plt.xlim(-10,11)

    fig = plt.gcf()
    fig.set_size_inches(10,6)
    plt.savefig('above_or_below_color_scheme.png')
    plt.show()

# Example usage
a = 100
b = 15
plot_function(a, b)

# Show the plot
plt.savefig('above_or_below_colour_scheme.png')
plt.show()
```
