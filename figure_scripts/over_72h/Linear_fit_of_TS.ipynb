```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Data points with x and y coordinates
x = [100, 87.5, 58.2, 65.25, 52.36, 33.02]
y = [1.416, 1.512, 0.847, 1.015, 0.799, 0.894]
err = [0.265, 0.285, 0.142, 0.172, 0.113, 0.136]

fig, ax = plt.subplots(figsize=[12,8])

# Scatter plot with individual marker styles and colors
marker_styles = ['o', 's', '^', 'o', 's', '^']
color_palette = cm.tab10.colors
color_palette2 = cm.Set2.colors
labels = ['Set A-high', 'Set A-medium', 'Set A-low', 'Set B-160°C', 'Set B-140°C', 'Set B-180°C']
colors = [color_palette[0], color_palette[1], color_palette[2], color_palette[0], color_palette[1], color_palette[2]]
#colors = ['black', 'green', 'red', 'black', 'green', 'red']
fill_styles = [color_palette[0], color_palette[1], color_palette[2], 'white', 'white', 'white']

for i in range(len(x)):
    ax.vlines(x[i], y[i] - err[i], y[i] + err[i], color='black', linewidth=2,  zorder=1)

for i in range(len(x)):
    ax.scatter(x[i], y[i], label=labels[i], edgecolor=colors[i], facecolors=fill_styles[i], marker=marker_styles[i], linewidth=3, s=120)


# Calculate linear fit
fit = np.polyfit(x, y, 1)
fit_fn = np.poly1d(fit)

# Generate x values for the line
x_line = np.linspace(min(x), max(x), 100)

# Plot linear fit line
ax.plot(x_line, fit_fn(x_line), '-', color='black', linewidth=2)


# Print the equation of the fitted line
slope = fit[0]
intercept = fit[1]
equation = f"y = {slope:.2f}x + {intercept:.2f}"
print("Equation of the fitted line:", equation)

# Set labels and title
legend = ax.legend(fontsize=16)
legend.set_frame_on(False)
ax.set_ylim([0.4, 1.9])
ax.set_xlim([23, 110])
ax.set_xlabel('Glucose conversion after 72h [%]', fontsize=18)
ax.set_ylabel('Normalised $S_{T}$ of crystallinity fraction of cellulose', fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=18)

text = '''
Linear Fit: y = 3.53E-1 + 1.1E-2*x
P-Value: 2.35E-2
Correlation Coefficient: 8.722E-1
'''.format(intercept, slope)
plt.text(0.95, 0.05, text, transform=plt.gca().transAxes, fontsize=16, ha='right', va='bottom')

# Show the plot
plt.savefig('TS_pct_cry_cell_over_72h.png')
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# Define your data
x = [100, 87.5, 58.2, 65.25, 52.36, 33.02]
y = [0.09, 0.23, 0.5, 0.41, 0.59, 0.72]

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(x, y)[0, 1]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Print the results
print(f"Correlation Coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")
print(f"Linear Fit (slope, intercept): ({slope}, {intercept})")
```
