```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.cm as cm
from matplotlib.colors import LinearSegmentedColormap
```

```python
param_values = np.loadtxt('param_values_trial_run.txt')
output_values = np.loadtxt('output_values_trial_run.txt')
color_values = np.loadtxt('above_or_below.txt')

param_names = ['EG_rxn_rate', 'CBH_rxn_rate', 'BGL_rxn_rate', 'XYL_rxn_rate', 
            'CBH_attachment_rate', 
            'binding_affinity_glucose_to_EG', 'binding_affinity_glucose_to_CBH', 'binding_affinity_glucose_to_BGL', 
            'init_EG', 'init_CBH', 'init_BGL', 'init_XYL',
            'pct_crystalline_cellu', 'pct_crystalline_hemi']
```

```python
i = 1
x_values = param_values[:, i]

print(x_values.shape)
print(output_values.shape)

color_palette = cm.tab10.colors

fig, ax = plt.subplots(figsize=[12,8])
for x, y, color in zip(x_values, output_values, color_values):
    if y <= 150:
        ax.plot(x, y, color=color_palette[0], marker='.', markersize=7)
    else:
        if color == 1:
            ax.plot(x, y, color=color_palette[8], marker='.', markersize=7)
        else:
            ax.plot(x, y, color=color_palette[9], marker='.', markersize=7)
#ax.set_xlim([0,0.01])
#ax.set_ylim([0, 3400])
ax.set_ylim([0,1200])
ax.set_xlabel('CBH reaction rate [1/h]', fontsize=23)
ax.set_ylabel('Y', fontsize=23)
#ax.set_title('Var over ' + param_names[i])
ax.tick_params(axis='both', which='major', labelsize=20)

plt.savefig('cloud_data_plots/04_06_var_vs_' + param_names[i] + '_color.png')
plt.show()
```

```python
i = 1
j = 9
x2_values = param_values[:, i]
color_data = param_values[:,j]


fig, ax = plt.subplots(figsize=[14,8])
#plt.figure(figsize=[8,8])
scatter = ax.scatter(x2_values, output_values, c=color_data, cmap='plasma')

#ax.set_xlim([0,1]
ax.set_ylim([0,1200])
ax.set_xlabel('CBH reaction rate [1/h]', fontsize=23)
ax.set_ylabel('Y', fontsize=23)
cbar = plt.colorbar(scatter)
cbar.set_label('Initial number of CBH enzymes', fontsize=23)
ax.tick_params(axis='both', which='major', labelsize=20)
#ax.set_title('Var over ' + param_names[1])
plt.savefig('cloud_data_plots/04_06_var_vs_' + param_names[i] + '_and_' + param_names[j])
plt.show()
```
