```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
from matplotlib.colors import LinearSegmentedColormap
```

```python
param_values = np.loadtxt('param_values_trial_run.txt')
output_values = np.loadtxt('output_values_trial_run.txt')
color_values = np.loadtxt('above_or_below.txt')

param_names = ['EG_rxn_rate', 'CBH_rxn_rate', 'BGL_rxn_rate', 'XYL_rxn_rate', 
            'Lignin_adhesion_rate', 'CBH_attachment_rate',
            'binding_affinity_cellobiose_to_EG', 'binding_affinity_cellobiose_to_CBH', 
            'binding_affinity_glucose_to_EG', 'binding_affinity_glucose_to_CBH', 'binding_affinity_glucose_to_BGL',
            'digestibility_ratio_cellulose', 'digestibility_ratio_hemicellulose',
            'init_EG', 'init_CBH', 'init_BGL', 'init_XYL',
            'pct_crystalline_cellu', 'pct_crystalline_hemi']
```

```python
i = 17
x_values = param_values[:, i]

print(x_values.shape)
print(output_values.shape)

#color_palette = list(mcolors.CSS4_COLORS.keys())[:10]
color_palette = cm.tab10.colors

fig, ax = plt.subplots(figsize=[12,8])
for x, y, color in zip(x_values, output_values, color_values):
    if y <= 150:
        ax.plot(x, y, color=color_palette[0], marker='.', markersize=7)
    else:
        if color == 1:
            ax.plot(x, y, color=color_palette[8], marker='.', markersize=7)
        else:
            ax.plot(x, y, color=color_palette[9], marker='.', markersize=7)
#ax.set_xlim([0,0.01])
#ax.set_ylim([0, 3400])
ax.set_ylim([0,5200])
ax.set_xlabel('Crystallinity fraction of cellulose', fontsize=24)
ax.set_ylabel('Y', fontsize=24)
ax.tick_params(axis='both', which='major', labelsize=22)

plt.savefig('cloud_data/07_05_var_vs_' + param_names[i] + '_color.png')
plt.show()
```
